```python
# Function 1: Perform basic mathematical operations
def calculate(a, b, op):
    if op == '+':
        return a + b          # Addition
    elif op == '-':
        return a - b          # Subtraction
    elif op == '*':
        return a * b          # Multiplication
    elif op == '/':
        # Check for division by zero
        return a / b if b != 0 else "Cannot divide by zero"
    elif op == '%':
        return a % b          # Modulus (remainder)
    else:
        return "Invalid operator"  # If user enters wrong operator


# Function 2: Calculate factorial of a number
def factorial(n):
    result = 1
    # Loop from 1 to n and multiply each number
    for i in range(1, n + 1):
        result *= i
    return result


# Function 3: Main function to control program flow
def main():
    # Display menu
    print("1. Basic Operations")
    print("2. Factorial")

    # Take user choice
    choice = input("Enter choice: ")

    if choice == '1':
        # Take two numbers (supports complex numbers like 2+3j)
        a = complex(input("Enter first number: "))
        b = complex(input("Enter second number: "))
        
        # Take operator input
        op = input("Enter operator (+, -, *, /, %): ")

        # Modulus does not work with complex numbers
        if op == '%':
            print("Modulo not supported for complex numbers")
        else:
            # Call calculate function and print result
            print("Result:", calculate(a, b, op))

    elif choice == '2':
        # Take integer input for factorial
        n = int(input("Enter number: "))
        
        # Call factorial function and print result
        print("Factorial:", factorial(n))

    else:
        # If user enters invalid choice
        print("Invalid choice")


# Run the program
main()
```
