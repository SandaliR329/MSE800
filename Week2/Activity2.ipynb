```python
# Global variable
students = []

# Student Class
class Student:
    def __init__(student1, name, age, student_id):
        # Local variables inside the class
        student1.name = name
        student1.age = age
        student1.student_id = student_id

# Function 1: Collect student data
def collect_students():
    for i in range(3):  # collect at least 3 students
        print(f"\nEnter details for student {i+1}")
        
        # Local variables
        name = input("Enter name: ")
        age = int(input("Enter age: "))
        student_id = input("Enter student ID: ")

        # Create Student object
        student = Student(name, age, student_id)

        # Add to global list
        students.append(student)


# Function 2: Display student names and ages
def display_students():
    print("\nStudent List (Name and Age):")
    
    # Sort students by name
    sorted_students = sorted(students, key=lambda s: s.age)
    
    for s in sorted_students:
        print(f"Name: {s.name}, Age: {s.age}")


# Entry point of the program
if __name__ == "__main__":
    collect_students()
    display_students()
```
